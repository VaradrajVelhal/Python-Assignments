#### **Question 3: Create a Python program that asks the user for the radius of a circle and then calculates and prints the area and circumference.**


```python
# Question 3: Create a Python program that asks the user for the radius of a circle and then calculates and prints the area and circumference.
# Write your code here
import math
radius = float(input("Enter radius of circle : "))
print("Radius you entered is : ", radius)
area = math.pi * radius * radius
circumference = 2 * math.pi * radius
print("Area of circle is : ", area)
print("Circumference of circle is : ", circumference)
```

#### **Question 4: Write a Python program that converts a temperature from Celsius to Fahrenheit and vice versa based on user input.**

```python
# Question 4: Write a Python program that converts a temperature from Celsius to Fahrenheit and vice versa based on user input.
# Write your code here
temp = float(input("Enter the temperature: "))
unit = input("Convert to Fahrenheit or Celsius? (F or C)").strip().upper()

if unit == 'F':
    result = (9/5) * temp + 32
    print(f"{temp} C is {result} F")
elif unit == 'C':
    result = (5/9) * (temp - 32)
    print(f"{temp} F is {result} C")
else:
    print("Invalid option. Please enter 'F' or 'C'.")

```

#### **Question 7: Write a Python program that takes a string as input and prints the number of characters, words, and sentences in the string.**

#### **Question 8: Develop a Python program that calculates simple interest. The user should input the principal, rate of interest, and time, and the program should output the interest amount.**

```python
# Question 8: Develop a Python program that calculates simple interest. The user should input the principal,
#  rate of interest, and time, and the program should output the interest amount.
# Write your code here
p = int(input("Enter the principle: "))
r = float(input("Enter the rate of interest: "))
t = int(input("Enter the time (In years): "))
interest = (p*r*t)/100
print(f"The Principle is: {p} Rate of interest is: {r}% Time is: {t} Years.")
print("Interest amount is: ", interest)
```

#### **Question 9: Write a Python program that takes a sentence as input and prints each word in reverse order while keeping the word sequence the same.**

```python
# Question 9: Write a Python program that takes a sentence as input and prints each word in reverse order while keeping the word sequence the same.
# Write your code here
sentence = input("Enter the sentence: ")
words = sentence.split()
reversedWords = [word [::-1] for word in words]
reversedWordsSentence = ' '.join(reversedWords)
print("The sentence you Entered is: ", sentence)
print("The reversed words sentence is: ",reversedWordsSentence)
```

#### **Question 10: Create a Python program that takes three numbers as input and prints the largest and smallest of the three.**

```python
# Question 10: Create a Python program that takes three numbers as input and prints the largest and smallest of the three.
# Write your code here
n1 = float(input("Enter First number : "))
n2 = float(input("Enter Second number : "))
n3 = float(input("Enter Third number : "))
print(f"The numbers you Entered are: 1st {n1} 2nd {n2} 3rd {n3} ")
if n1 == n2 and n2 == n3:
    print("All are equal....")
else : 
 largest = max(n1, n2, n3)
 smallest = min(n1, n2, n3)
 print("The largest number is : ", largest)
 print("Smallest number is :", smallest)
```
