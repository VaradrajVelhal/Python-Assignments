**1.Create a program that models a basic shape hierarchy. Define a base class Shape
with methods for calculating area and perimeter. Create subclasses for Circle and
Rectangle.**


```python
class Shape:
    def area(self):
        pass

    def perimeter(self):
        pass


class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius * self.radius

    def perimeter(self):
        return 2 * 3.14 * self.radius


class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * (self.length + self.width)


circle = Circle(7)
rectangle = Rectangle(5, 3)

print("Circle Area:", circle.area())
print("Circle Perimeter:", circle.perimeter())

print("Rectangle Area:", rectangle.area())
print("Rectangle Perimeter:", rectangle.perimeter())
```

with a method draw(). Create subclasses Circle and Square that implement the
draw() method.**



```python
class Shape:
    def draw(self):
        print("Drawing a shape")


class Circle(Shape):
    def draw(self):
        print("Drawing a circle")


class Square(Shape):
    def draw(self):
        print("Drawing a square")


shapes = [Circle(), Square()]

for shape in shapes:
    shape.draw()
```

shopping cart. Define a base class Item and subclasses for different types of items.**


```python
class Item:
    def get_price(self):
        return 0


class Book(Item):
    def __init__(self, price):
        self.price = price

    def get_price(self):
        return self.price


class Electronic(Item):
    def __init__(self, price):
        self.price = price

    def get_price(self):
        return self.price


cart = [
    Book(300),
    Electronic(1500),
    Book(200)
]

total = 0
for item in cart:
    total += item.get_price()

print("Total cost:", total)
```

methods for calculating area and perimeter. Create subclasses Circle and
Rectangle to implement these methods.**



```python
from abc import ABC, abstractmethod

class Shape(ABC):

    @abstractmethod
    def area(self):
        pass

    @abstractmethod
    def perimeter(self):
        pass


class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius * self.radius

    def perimeter(self):
        return 2 * 3.14 * self.radius


class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * (self.length + self.width)


circle = Circle(7)
rectangle = Rectangle(5, 3)

print("Circle Area:", circle.area())
print("Circle Perimeter:", circle.perimeter())

print("Rectangle Area:", rectangle.area())
print("Rectangle Perimeter:", rectangle.perimeter())
```

payments. Implement the interface in classes CreditCardPayment and
PayPalPayment**


```python
from abc import ABC, abstractmethod

class Payment(ABC):

    @abstractmethod
    def process_payment(self, amount):
        pass


class CreditCardPayment(Payment):
    def process_payment(self, amount):
        print("Credit Card payment of", amount, "processed")


class PayPalPayment(Payment):
    def process_payment(self, amount):
        print("PayPal payment of", amount, "processed")


payment1 = CreditCardPayment()
payment2 = PayPalPayment()

payment1.process_payment(2000)
payment2.process_payment(1500)
```


**6.Write a Python program that models a simple bank account hierarchy. Create a
base class Account with attributes for account_number, account_holder, and
balance. Create subclasses for SavingsAccount and CurrentAccount..**




```python
class Account:
    def __init__(self, account_number, account_holder, balance):
        self.account_number = account_number
        self.account_holder = account_holder
        self.balance = balance

    def display_details(self):
        print("Account Number:", self.account_number)
        print("Account Holder:", self.account_holder)
        print("Balance:", self.balance)


class SavingsAccount(Account):
    def __init__(self, account_number, account_holder, balance, interest_rate):
        super().__init__(account_number, account_holder, balance)
        self.interest_rate = interest_rate

    def display_details(self):
        super().display_details()
        print("Interest Rate:", self.interest_rate)


class CurrentAccount(Account):
    def __init__(self, account_number, account_holder, balance, overdraft_limit):
        super().__init__(account_number, account_holder, balance)
        self.overdraft_limit = overdraft_limit

    def display_details(self):
        super().display_details()
        print("Overdraft Limit:", self.overdraft_limit)


savings = SavingsAccount(101, "Varadraj", 5000, 4.5)
current = CurrentAccount(102, "Rahul", 8000, 10000)

print("Savings Account Details")
savings.display_details()

print("\nCurrent Account Details")
current.display_details()
```

for name and sound. Create subclasses for different types of animals (e.g., Lion,
Elephant) that inherit from the Animal class..**


 

```python
class Animal:
    def __init__(self, name, sound):
        self.name = name
        self.sound = sound

    def make_sound(self):
        print(self.name, "says", self.sound)


class Lion(Animal):
    def __init__(self, name):
        super().__init__(name, "Roar")


class Elephant(Animal):
    def __init__(self, name):
        super().__init__(name, "Trumpet")


lion = Lion("Simba")
elephant = Elephant("Dumbo")

lion.make_sound()
elephant.make_sound()
```

Define a base class Course with methods for enrolling and dropping students.
Create subclasses for different types of courses (e.g., MathCourse,
HistoryCourse).**


```python
class Course:
    def __init__(self, course_name):
        self.course_name = course_name
        self.students = []

    def enroll_student(self, student_name):
        self.students.append(student_name)
        print(student_name, "enrolled in", self.course_name)

    def drop_student(self, student_name):
        if student_name in self.students:
            self.students.remove(student_name)
            print(student_name, "dropped from", self.course_name)
        else:
            print(student_name, "not found in", self.course_name)


class MathCourse(Course):
    def __init__(self):
        super().__init__("Mathematics")


class HistoryCourse(Course):
    def __init__(self):
        super().__init__("History")


math = MathCourse()
history = HistoryCourse()

math.enroll_student("Varadraj")
math.enroll_student("Rahul")
math.drop_student("Rahul")

history.enroll_student("Amit")
```

DrawingTool with methods for drawing shapes. Create subclasses for different
types of drawing tools (e.g., PenTool, BrushTool) with specific drawing behavior.**

```python
class DrawingTool:
    def draw(self, shape):
        print("Drawing a shape")


class PenTool(DrawingTool):
    def draw(self, shape):
        print("Drawing", shape, "using Pen tool")


class BrushTool(DrawingTool):
    def draw(self, shape):
        print("Drawing", shape, "using Brush tool")


pen = PenTool()
brush = BrushTool()

pen.draw("Circle")
brush.draw("Rectangle")
```
