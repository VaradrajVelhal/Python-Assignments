```python
filename = "example.txt"
text_to_append = input("Enter text to append: ")

try:
    with open(filename, "a", encoding="utf-8") as file:
        file.write(text_to_append + "\n")
    
    with open(filename, "r", encoding="utf-8") as file:
        content = file.read()
        print("\nUpdated file contents:\n")
        print(content)

except FileNotFoundError:
    print(f"Error: {filename} not found.")
except Exception as e:
    print("An error occurred:", e)


```

```python
filename = "example.txt"

try:
    lines = []
    with open(filename, "r", encoding="utf-8") as file:
        for line in file:
            lines.append(line.strip())

    print("Lines in the file as a list:")
    print(lines)

except FileNotFoundError:
    print(f"Error: {filename} not found.")
except Exception as e:
    print("An error occurred:", e)

```

```python
import re

filename = "example.txt"

try:
    with open(filename, "r", encoding="utf-8") as file:
        text = file.read()

    words = re.findall(r'\b\w+\b', text)
    max_length = max(len(word) for word in words)
    longest_words = [word for word in words if len(word) == max_length]

    print("Longest word length:", max_length)
    print("Longest word:", longest_words)

except FileNotFoundError:
    print(f"Error: {filename} not found.")
except Exception as e:
    print("An error occurred:", e)
```

```python
import string

filename = "alphabet.txt"
letters_per_line = int(input("Enter the number of letters per line: "))

alphabet = string.ascii_lowercase

with open(filename, "w", encoding="utf-8") as file:
    for i in range(0, len(alphabet), letters_per_line):
        line = alphabet[i:i+letters_per_line]
        file.write(line + "\n")

print(f"\nAlphabet written to {filename}\n")
print("File contents:\n")

with open(filename, "r", encoding="utf-8") as file:
    content = file.read()
    print(content)
```


**7.Write code to read a CSV file containing product information and create a list of product objects.**


 

```python
import csv

class Product:
    def __init__(self, name, price, quantity):
        self.name = name
        self.price = float(price)
        self.quantity = int(quantity)
    def __repr__(self):
        return f"Product(name={self.name}, price={self.price}, quantity={self.quantity})"

filename = "products.csv"

# Create CSV file
with open(filename, "w", newline='', encoding="utf-8") as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(["name", "price", "quantity"])
    writer.writerow(["Laptop", 1200.50, 5])
    writer.writerow(["Mouse", 25.99, 50])
    writer.writerow(["Keyboard", 45.00, 30])

print(f"Contents of {filename}:")
with open(filename, "r", encoding="utf-8") as csvfile:
    print(csvfile.read())

# Read CSV into product objects
products = []
with open(filename, newline='', encoding="utf-8") as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        products.append(Product(row['name'], row['price'], row['quantity']))

# Print product objects
print("\nProduct objects:")
for p in products:
    print(p)
```

```python
filename = "example.txt"

try:
    with open(filename, "r", encoding="utf-8") as file:
        text = file.read()

    lines = text.splitlines()
    num_lines = len(lines)
    num_words = len(text.split())
    num_chars = len(text)

    print(f"Number of lines: {num_lines}")
    print(f"Number of words: {num_words}")
    print(f"Number of characters: {num_chars}")

except FileNotFoundError:
    print(f"Error: {filename} not found.")
except Exception as e:
    print("An error occurred:", e)
```

```python
filename = "user_input.txt"

try:
    with open(filename, "w", encoding="utf-8") as file:
        while True:
            line = input("Enter a line (or type 'STOP' to finish): ")
            if line.upper() == "STOP":
                break
            file.write(line + "\n")
    print(f"Lines have been written to {filename}")

    with open(filename, "r", encoding="utf-8") as file:
        print("\nContents of the file:")
        print(file.read())

except Exception as e:
    print("An error occurred:", e)
```
