#### **Question 7: List of prime numbers up to N using math.isqrt().**

```python
import math

n = input("Enter limit: ")
print(f"Inputs are {n}")

n = int(n)
primes = []

for x in range(2, n+1):
    f = True
    i = 2
    while i <= math.isqrt(x):
        if x % i == 0:
            f = False
            break
        i = i + 1
    if f:
        primes.append(x)

print(primes)
```
