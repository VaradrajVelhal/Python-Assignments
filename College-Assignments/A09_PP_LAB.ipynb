within multiple layers of function calls.**


```python
class NegativeNumberError(Exception):
    pass

def divide(a, b):
    try:
        if b < 0:
            raise NegativeNumberError("Denominator cannot be negative")
        return a / b
    except ZeroDivisionError:
        print("Error: Division by zero in divide()")
        return None

def calculate():
    try:
        x = int(input("Enter numerator: "))
        y = int(input("Enter denominator: "))
        result = divide(x, y)
        if result is not None:
            print("Result of division:", result)
    except ValueError:
        print("Error: Invalid input in calculate()")

try:
    calculate()
except NegativeNumberError as e:
    print("Caught custom exception in main:", e)
```

regardless of exceptions or successful execution.**



```python
try:
    file = open("example.txt", "r")
    data = file.read()
    print("File contents:")
    print(data)
except FileNotFoundError:
    print("Error: File not found")
except IOError:
    print("Error: An I/O error occurred")
else:
    print("File read successfully")
finally:
    if 'file' in locals() and not file.closed:
        file.close()
        print("File has been closed (resource released)")
```

operations in case of temporary failures.**


```python
import random
import time

class TemporaryError(Exception):
    pass

def unstable_operation():
    if random.choice([True, False]):
        raise TemporaryError("Temporary failure occurred")
    return "Operation succeeded"

def retry_operation(max_retries=3, delay=2):
    attempts = 0
    while attempts < max_retries:
        try:
            result = unstable_operation()
            print(result)
            break
        except TemporaryError as e:
            attempts += 1
            print(f"{e} — retrying {attempts}/{max_retries}...")
            time.sleep(delay)
    else:
        print("Operation failed after maximum retries")

retry_operation()
```

exception and print a message if the program does not have permission to read the
file.**

prints the average of those numbers. Handle the ValueError exception when
converting user input to numbers.**



```python
numbers = []

while True:
    try:
        nums = input("Enter numbers separated by spaces: ").split()
        numbers = [float(num) for num in nums]
        break
    except ValueError:
        print("Invalid input. Please enter only numbers.")

if numbers:
    print("Numbers entered:", numbers)
    average = sum(numbers) / len(numbers)
    print("Average of the numbers:", average)
else:
    print("No numbers were entered.")
```

mathematical operation (e.g., addition, subtraction, multiplication, division) based on
user input Handle the potential exceptions that may arise during user input or
calculation.**


```python
try:
    num1 = float(input("Enter the first number: "))
    num2 = float(input("Enter the second number: "))
    
    operation = input("Choose operation (+, -, *, /): ")

    if operation == "+":
        result = num1 + num2
    elif operation == "-":
        result = num1 - num2
    elif operation == "*":
        result = num1 * num2
    elif operation == "/":
        result = num1 / num2
    else:
        raise ValueError("Invalid operation")

except ValueError as e:
    print("Error:", e)
except ZeroDivisionError:
    print("Error: Cannot divide by zero")
else:
    print(f"The result of {num1} {operation} {num2} is: {result}")
finally:
    print("Calculation attempt completed.")
```

password. Handle the KeyError when checking if the entered username exists in a
dictionary and handle the ValueError when comparing the entered password.**

```python
users = {
    "Mahesh": "password123",
    "Kamlesh": "Kamlesh123",
    "Ganesh": "Ganesh123"
}

try:
    username = input("Enter username: ")
    password = input("Enter password: ")

    if username not in users:
        raise KeyError(f"Username '{username}' does not exist")

    if users[username] != password:
        raise ValueError("Incorrect password")

except KeyError as e:
    print("Login error:", e)
except ValueError as e:
    print("Login error:", e)
else:
    print(f"Login successful. Welcome, {username}!")
finally:
    print("Login attempt completed.")
```
