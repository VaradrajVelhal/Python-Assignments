#### **Question 1: Work with Dates (datetime)**


```python
import datetime

d = input("Enter date as YYYY-MM-DD: ")
print(f"Inputs are {d}")
y, m, day = [int(x) for x in d.split("-")]
dt = datetime.date(y, m, day)
print(dt)
print(dt.weekday())
```

#### **Question 2: Generate Random Numbers (random).**

```python
import random

n = input("How many random numbers? ")
print(f"Inputs are {n}")
n = int(n)
for i in range(n):
    print(random.random())
```

#### **Question 3: Display Calendar of a Month (calendar).**

```python
import calendar

y = input("Enter year: ")
print(f"Inputs are {y}")
m = input("Enter month number: ")
print(f"Inputs are {m}")
y = int(y)
m = int(m)
print(calendar.month(y, m))
```

#### **Question 4: Area and Circumference of a Circle.**

```python
import math

r = input("Enter radius: ")
print(f"Inputs are {r}")
r = float(r)
area = math.pi * r * r
circ = 2 * math.pi * r
print(area)
print(circ)
```

#### **Question 5: Generate a Random Number Between Two Numbers.**

#### **Question 7: Create a Directory and List Files in it (os).**

```python
import os

name = input("Enter directory name: ")
print(f"Inputs are {name}")
if not os.path.exists(name):
    os.mkdir(name)
print(os.listdir(name))
```

#### **Question 8: Shuffle a List of Numbers (random).**

```python
import random

s = input("Enter numbers separated by comma: ")
print(f"Inputs are {s}")
first = [x.strip() for x in s.split(",") if x.strip() != ""]
random.shuffle(first)
print(first)
```
