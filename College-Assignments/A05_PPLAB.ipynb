#### **Question 1: Calculate area of rectangle using default arguments**


```python
def area(w=1, h=1):
    return w * h

w = input("Enter width: ")
print(f"Inputs for width is: {w}")
h = input("Enter height: ")
print(f"Inputs for height is: {h}")

if w == "":
    w = 1
else:
    w = float(w)

if h == "":
    h = 1
else:
    h = float(h)

print(area(w, h))
print(area())
```

#### **Question 3: Product of Any Number of Arguments.**

```python
vals = input("Enter numbers separated by spaces: ")
print(f"Inputs are {vals}")

parts = vals.split()
parts = [float(x) for x in parts]

p = 1
for x in parts:
    p = p * x

print(p)
```

#### **Question 4: Sort Tuples by Second Element using lambda.**

```python
n = input("Enter number of tuples: ")
print(f"Inputs are {n}")
n = int(n)

t = []
for i in range(n):
    x = input("Enter tuple like a,3: ")
    print(f"Inputs are {x}")
    a, b = x.split(",")
    t.append((a, float(b)))

t = sorted(t, key=lambda x: x[1])
print(t)
```

#### **Question 5: Factorial Using Recursion.**

```python
n = input("Enter number: ")
print(f"Inputs are {n}")
n = int(n)

def fact(x):
    if x == 0:
        return 1
    else:
        return x * fact(x-1)

print(fact(n))
```

#### **Question 6: Square Each Element in a List.**

```python
nums = input("Enter numbers separated by commas: ")
print(f"Inputs are {nums}")

lst = nums.split(",")
lst = [float(x) for x in lst]

new = []
for x in lst:
    new.append(x*x)

print(new)
```

#### **Question 7: Convert Strings to Uppercase.**

```python
s = input("Enter strings separated by commas: ")
print(f"Inputs are {s}")

lst = s.split(",")
lst = [x.strip() for x in lst]

for i in range(len(lst)):
    lst[i] = lst[i].upper()

print(lst)
```

#### **Question 8: Check If Number Is Prime.**

```python
n = input("Enter number: ")
print(f"Inputs are {n}")
n = int(n)

if n <= 1:
    print(False)
else:
    flag = True
    i = 2
    while i < n:
        if n % i == 0:
            flag = False
            break
        i = i + 1
    print(flag)
```

#### **Question 9: Print Dictionary Key: Value.**

```python
d = input("Enter key:value pairs separated by colan: ")
print(f"Inputs are {d}")

pairs = d.split(",")
dic = {}

for p in pairs:
    k, v = p.split(":")
    dic[k] = v

for k in dic:
    print(k + ": " + dic[k])
```

#### **Question 10: Nested Function.**

```python
a = input("Enter first number: ")
print(f"Inputs are {a}")
b = input("Enter second number: ")
print(f"Inputs are {b}")

a = float(a)
b = float(b)

def outer(x, y):
    def inner(m, n):
        return m + n
    return inner(x, y)

print(outer(a, b))
```
