#### **Question 1: Design a Python program to simulate a simple game of rock-paper-scissors.**

```python
# Question 1: Design a Python program to simulate a simple game of rock-paper
# -scissors.
# Write your code here
import random

choices = ('rock', 'paper', 'scissors')

wins_against = {
    'rock': 'scissors',
    'paper': 'rock',
    'scissors': 'paper'
}

user = input("Choose rock, paper, or scissors: ").lower()
print(f"User choice: {user}")

if user not in choices:
    print("Invalid choice!")
else:
    computer = random.choice(choices)
    print("Computer chose:", computer)

    if user == computer:
        print("It's a tie!")
    elif wins_against[user] == computer:
        print("You win!")
    else:
        print("Computer wins!")
```

#### **Question 2: Develop a Python code to generate a random password with specified criteria.**

```python
# Question 2: Develop a Python code to generate a random password with
#  specified criteria.
# Write your code here
import random
import string

chars = string.ascii_letters + string.digits + string.punctuation

length = input("Enter the password length: ")

if length.isdigit() and int(length) > 0:
    length = int(length)
    password = ''
    for _ in range(length):
        password += random.choice(chars)
    print(f"Generated password for length {length} : {password}")
else:
    print("Please enter a valid positive number.")
```

#### **Question 3: Create a Python script that calculates the area and perimeter of different geometric shapes based on user-provided dimensions.**


```python
# Question 3: Create a Python script that calculates the area and perimeter
#  of different geometric shapes based on user-provided dimensions.
# Write your code here
import math

shapes = {
    'rectangle': ('length', 'width'),
    'circle': ('radius',)
}

shape = input("Choose shape (rectangle, circle): ").lower()

if shape not in shapes:
    print("Unsupported shape!")
else:
    print("You entered :", shape)
    dimensions = {}
    for dim in shapes[shape]:
        dimensions[dim] = float(input(f"Enter {dim}: "))

    if shape == 'rectangle':
        length, width = dimensions['length'], dimensions['width']
        area = length * width
        perimeter = 2 * (length + width)
    else:
        radius = dimensions['radius']
        area = math.pi * radius**2
        perimeter = 2 * math.pi * radius

    print(f"Area: {area:.2f}")
    print(f"Perimeter: {perimeter:.2f}")

```

#### **Question 5: Create a Python script that converts a Fahrenheit temperature to Celsius and vice versa, using appropriate data types and conversion formulas. **

```python
# Question 5: Create a Python script that converts a Fahrenheit
# temperature to Celsius and vice versa, using appropriate data
#  types and conversion formulas. 
# Write your code here
temp_data = {}

choice = input("Type 'F' to convert Fahrenheit to Celsius or 'C' for Celsius to Fahrenheit: ").upper()

if choice == 'F':
    temp_data['Fahrenheit'] = float(input("Enter temperature in Fahrenheit: "))
    # Convert to Celsius
    temp_data['Celsius'] = (temp_data['Fahrenheit'] - 32) * 5 / 9
    print(f"{temp_data['Fahrenheit']}°F is {temp_data['Celsius']:.2f}°C")

elif choice == 'C':
    temp_data['Celsius'] = float(input("Enter temperature in Celsius: "))
    # Convert to Fahrenheit
    temp_data['Fahrenheit'] = (temp_data['Celsius'] * 9 / 5) + 32
    print(f"{temp_data['Celsius']}°C is {temp_data['Fahrenheit']:.2f}°F")

else:
    print("Invalid choice! Please type 'F' or 'C'.")
```

#### **Question 6: Implement a Python function to sort a list of numbers using the bubble sort algorithm. **

```python
# Question 6: Implement a Python function to sort a list of
#  numbers using the bubble sort algorithm. 
# Write your code here

def bubbleSort(numbers):
    n = len(numbers)
    for i in range(n):
        for j in range(0, n - i - 1):
            if numbers[j] > numbers[j + 1]:
                numbers[j], numbers[j + 1] = numbers[j + 1], numbers[j]
    return numbers

nums = [12, 11, 8, 6, 32, 10]
sorted_nums = bubbleSort(nums)
print("sorted numbers are:", sorted_nums)




```

#### **Question 7: Write a Python program that takes a list of numbers as input and prints the sum of all the even numbers in the list**

```python
# Question 7:Write a Python program that takes a list of numbers as input and
#  prints the sum of all the even numbers in the list.
# Write your code here

# Function to count sum of even numbers
def sum_of_even_nums(numbers):
    total = 0
    for num in numbers:
        if num % 2 == 0:
            total += num
    return total

input_numbers = input("Enter numbers separated by spaces: ").split()

# Converting the input strings to integers
numbers_list = [int(num) for num in input_numbers]

result = sum_of_even_nums(numbers_list)

print("The numbers you entered are : ", input_numbers)

print("Sum of even numbers:", result)
```

#### **Question 8: Write a Python program that takes a list of strings and prints the count of each unique word in the list. **

```python
# Question 8: Write a Python program that takes a list of strings 
# and prints the count of each unique word in the list. 
# Write your code here

# Function to count occurrences of unique word
def count_words(words):
    word_count = {}
    for word in words:
        if word in word_count:
            word_count[word] += 1
        else:
            word_count[word] = 1
    return word_count

input_words = input("Enter words separated by spaces: ").split()

# Counting the words
counts = count_words(input_words)

print("The words you entered are : ", input_words)


# Printing the counts
for word, count in counts.items():
    print(f"'{word}': {count}")
```

#### **Question 9: Write a Python program that defines two tuples of numbers and calculates the element-wise sum of the tuples. **

```python
# Question 9: Write a Python program that defines two tuples of
# numbers and calculates the element-wise sum of the tuples. 
# Write your code here

tuple1 = (1, 2, 3)
tuple2 = (4, 5, 6)

# Calculating element wise sum.
elementwise_sum = tuple(a + b for a, b in zip(tuple1, tuple2))

# Print the result
print("Element-wise sum:", elementwise_sum)
```
