**1.Write a Python program that connects to a MySQL database, updates the salary of
an employee in the employees table, and then queries and prints the updated
record**


```python
# Question 1: 
import mysql.connector

# connect to MySQL database
conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="2910",
    database="college"
)

cursor = conn.cursor()

# update salary of an employee
update_query = "UPDATE employees SET salary = %s WHERE name = %s"
cursor.execute(update_query, (45000, "Varadraj"))
conn.commit()

# fetch and display updated record
cursor.execute("SELECT name, salary FROM employees WHERE name = 'Varadraj'")
record = cursor.fetchone()

print("Updated Employee Record:")
print("Name:", record[0])
print("Salary:", record[1])

conn.close()
```

MySQL database, start a transaction, insert multiple records into a table, and
commit the transaction.**


```python
# Question 3:

import mysql.connector

# connect to MySQL database
conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="2910",
    database="college"
)

cursor = conn.cursor()

try:
    # start transaction
    conn.start_transaction()

    # insert multiple records
    cursor.execute(
        "INSERT INTO employees (name, salary) VALUES (%s, %s)",
        ("Amit", 30000)
    )
    cursor.execute(
        "INSERT INTO employees (name, salary) VALUES (%s, %s)",
        ("Rahul", 35000)
    )
    cursor.execute(
        "INSERT INTO employees (name, salary) VALUES (%s, %s)",
        ("Varadraj", 40000)
    )

    # commit transaction
    conn.commit()
    print("Transaction committed successfully")

except Exception as e:
    # rollback if any error occurs
    conn.rollback()
    print("Transaction failed, rollback done")

conn.close()
```

containing student grades, and calculates and prints the average grade..**


```python
import mysql.connector

try:
    # Connect to MySQL database
    conn = mysql.connector.connect(
        host="localhost",
        user="root",
        password="2910",
        database="college"
    )

    cursor = conn.cursor()

    # Query grades
    cursor.execute("SELECT grade FROM students")

    grades = cursor.fetchall()

    total = 0
    count = 0

    # Calculate the sum and count, skipping None values
    for g in grades:
        if g[0] is not None:  # Check if the grade is not None
            try:
                total += float(g[0])  # Ensure grade is treated as a number
                count += 1
            except ValueError:
                print(f"Invalid grade value: {g[0]}")

    if count > 0:
        average = total / count
        print("Average Grade:", average)
    else:
        print("No records found")

except mysql.connector.Error as err:
    print(f"Error: {err}")
finally:
    # Clean up
    if cursor:
        cursor.close()
    if conn:
        conn.close()
```


**7.Write a Python program that connects to a MySQL database and uses a prepared
statement to insert multiple records into a table.**


 

```python
# Question 7:
import mysql.connector

# connect to MySQL database
conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="2910",
    database="college"
)

cursor = conn.cursor()

# prepared statement
sql = "INSERT INTO employees (name, salary) VALUES (%s, %s)"

# multiple records
records = [
    ("Amit", 30000),
    ("Rahul", 35000),
    ("Varadraj", 40000)
]

# execute prepared statement for multiple records
cursor.executemany(sql, records)

conn.commit()

print("Multiple records inserted successfully")

conn.close()

```

named students and courses, establishes a relationship between them, and inserts
sample data**


```python
# Question 9:

import mysql.connector

# connect to MySQL database
conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="2910",
    database="college"
)

cursor = conn.cursor()

# create students table
cursor.execute("""
CREATE TABLE IF NOT EXISTS students (
    student_id INT AUTO_INCREMENT PRIMARY KEY,
    name VARCHAR(50)
)
""")

# create courses table with foreign key
cursor.execute("""
CREATE TABLE IF NOT EXISTS courses (
    course_id INT AUTO_INCREMENT PRIMARY KEY,
    course_name VARCHAR(50),
    student_id INT,
    FOREIGN KEY (student_id) REFERENCES students(student_id)
)
""")

# insert sample students
cursor.execute("INSERT INTO students (name) VALUES (%s)", ("Varadraj",))
cursor.execute("INSERT INTO students (name) VALUES (%s)", ("Rahul",))

# insert sample courses
cursor.execute(
    "INSERT INTO courses (course_name, student_id) VALUES (%s, %s)",
    ("Python", 1)
)
cursor.execute(
    "INSERT INTO courses (course_name, student_id) VALUES (%s, %s)",
    ("Database Systems", 2)
)

conn.commit()

print("Tables created and sample data inserted successfully")

conn.close()
```

on the students and courses tables, and prints the names of students along with the
courses they are enrolled in**

```python
import mysql.connector

try:
    # Connect to MySQL database
    conn = mysql.connector.connect(
        host="localhost",
        user="root",
        password="2910",
        database="college"
    )

    cursor = conn.cursor()

    # JOIN query
    query = """
    SELECT students.name, courses.course_name
    FROM students
    JOIN courses
    ON students.id = courses.student_id  -- Adjusted column names
    """

    cursor.execute(query)

    print("Student Enrollments:")
    # Fetch and print results
    for row in cursor.fetchall():
        print("Student:", row[0], "| Course:", row[1])

except mysql.connector.Error as err:
    print(f"Error: {err}")

finally:
    # Clean up
    if cursor:
        cursor.close()
    if conn:
        conn.close()
```
