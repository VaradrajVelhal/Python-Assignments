**1.Write a Python program to create a table and insert some records in that table.
Finally selects all rows from the table and display the records
.**


```python
import sqlite3

# connect to database (creates file if not exists)
conn = sqlite3.connect("student.db")
cursor = conn.cursor()

# create table
cursor.execute("""
CREATE TABLE IF NOT EXISTS Student (
    id INTEGER PRIMARY KEY,
    name TEXT,
    age INTEGER
)
""")

# insert records
cursor.execute("INSERT INTO Student (name, age) VALUES ('Varadraj', 22)")
cursor.execute("INSERT INTO Student (name, age) VALUES ('Rahul', 21)")
cursor.execute("INSERT INTO Student (name, age) VALUES ('Amit', 23)")

conn.commit()

# select and display records
cursor.execute("SELECT * FROM Student")
rows = cursor.fetchall()

print("Student Records:")
for row in rows:
    print(row)

# close connection
conn.close()
```

select all rows before and after updating the said table.**



```python
import sqlite3

# connect to database
conn = sqlite3.connect("student.db")
cursor = conn.cursor()

# create table
cursor.execute("""
CREATE TABLE IF NOT EXISTS Student (
    id INTEGER PRIMARY KEY,
    name TEXT,
    age INTEGER
)
""")

# insert records (only once, safe for exams)
cursor.execute("INSERT INTO Student (name, age) VALUES ('Varadraj', 22)")
cursor.execute("INSERT INTO Student (name, age) VALUES ('Rahul', 21)")
conn.commit()

# display records BEFORE update
print("Records before update:")
cursor.execute("SELECT * FROM Student")
rows = cursor.fetchall()
for row in rows:
    print(row)

# update age of Rahul
cursor.execute("UPDATE Student SET age = 23 WHERE name = 'Rahul'")
conn.commit()

# display records AFTER update
print("\nRecords after update:")
cursor.execute("SELECT * FROM Student")
rows = cursor.fetchall()
for row in rows:
    print(row)

# close connection
conn.close()
```

```python
import sqlite3

# connect to database
conn = sqlite3.connect("student.db")
cursor = conn.cursor()

# create table
cursor.execute("""
CREATE TABLE IF NOT EXISTS Student (
    id INTEGER PRIMARY KEY,
    name TEXT,
    age INTEGER
)
""")

# insert sample records
cursor.execute("INSERT INTO Student (name, age) VALUES ('Varadraj', 22)")
cursor.execute("INSERT INTO Student (name, age) VALUES ('Rahul', 21)")
conn.commit()

# display records BEFORE delete
print("Records before delete:")
cursor.execute("SELECT * FROM Student")
for row in cursor.fetchall():
    print(row)

# delete record where name is Rahul
cursor.execute("DELETE FROM Student WHERE name = 'Rahul'")
conn.commit()

# display records AFTER delete
print("\nRecords after delete:")
cursor.execute("SELECT * FROM Student")
for row in cursor.fetchall():
    print(row)

# close connection
conn.close()
```

```python
import sqlite3

# connect to database
conn = sqlite3.connect("student.db")
cursor = conn.cursor()

# create table
cursor.execute("""
CREATE TABLE IF NOT EXISTS Student (
    id INTEGER PRIMARY KEY,
    name TEXT,
    age INTEGER
)
""")

# alter table: add new column
cursor.execute("ALTER TABLE Student ADD COLUMN grade TEXT")

conn.commit()

print("Table altered successfully. New column 'grade' added.")

# display table data
cursor.execute("SELECT * FROM Student")
rows = cursor.fetchall()

print("Student Records:")
for row in rows:
    print(row)

# close connection
conn.close()
```


**6.Write Python code to:
a. Connect to a MySQL database using the mysql.connector library.
b. Execute a SQL query to retrieve all customer names and emails.
c. Insert a new row into the "customers" table.
d. Update an existing customer's email address**




```python
import sqlite3

# connect to database
conn = sqlite3.connect("customers.db")
cursor = conn.cursor()

# create table
cursor.execute("""
CREATE TABLE IF NOT EXISTS customers (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT,
    email TEXT
)
""")

# insert a new row
cursor.execute("INSERT INTO customers (name, email) VALUES (?, ?)",
               ("Varadraj", "varadraj@gmail.com"))
conn.commit()

# update email
cursor.execute("UPDATE customers SET email = ? WHERE name = ?",
               ("varadraj_new@gmail.com", "Varadraj"))
conn.commit()

# retrieve all customer names and emails
cursor.execute("SELECT name, email FROM customers")
rows = cursor.fetchall()

print("Customer Records:")
for row in rows:
    print(row)

conn.close()
```

named employees with columns for id, name, and salary, and inserts a few
records into the table.**


```python
# Question 9:
import mysql.connector

# connect to MySQL server
conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="2910",
    database="college"
)

cursor = conn.cursor()

# create table
cursor.execute("""
CREATE TABLE IF NOT EXISTS employees (
    id INT AUTO_INCREMENT PRIMARY KEY,
    name VARCHAR(50),
    salary INT
)
""")

# insert records
cursor.execute("INSERT INTO employees (name, salary) VALUES (%s, %s)", ("Amit", 30000))
cursor.execute("INSERT INTO employees (name, salary) VALUES (%s, %s)", ("Rahul", 35000))
cursor.execute("INSERT INTO employees (name, salary) VALUES (%s, %s)", ("Varadraj", 40000))

conn.commit()

print("Employees table created and records inserted successfully")

conn.close()

```

```python
# Question 10:



import mysql.connector

# connect to MySQL database
conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="2910",
    database="college"
)

cursor = conn.cursor()

# query employees table
cursor.execute("SELECT name, salary FROM employees")

print("Employee Details:")
for row in cursor.fetchall():
    print("Name:", row[0], "| Salary:", row[1])

conn.close()

```
