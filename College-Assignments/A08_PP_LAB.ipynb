.**


```python
class PythonException(Exception):
    pass

def checkValue(x):
    if x < 0:
        raise PythonException("Negative value not allowed")
    return x

try:
    a = int(input("Enter the number: "))
    checkValue(a)
    print("You entered:", a)
except PythonException as e:
    print(e)
```


**7.Create a custom exception class called NegativeNumberError to signal invalid input.**


 

```python
class NegativeNumberError(Exception):
    pass

def check_number(n):
    if n < 0:
        raise NegativeNumberError("Negative numbers are not allowed")
    return n

try:
    x = int(input("Enter a number: "))
    check_number(x)
    print("Valid number:", x)
except NegativeNumberError as e:
    print(e)
```



**8. Write a function that raises a custom exception if a provided argument is not a valid
file path.**



```python
import os

class InvalidFilePathError(Exception):
    pass

def check_file_path(path):
    if not os.path.isfile(path):
        raise InvalidFilePathError("Invalid file path: " + path)
    return path

try:
    p = input("Enter file path: ")
    check_file_path(p)
    print("Valid file path:", p)
except InvalidFilePathError as e:
    print(e)
```


**9.Write a function that raises an InsufficientFundsError if a withdrawal amount exceeds the available balance.**


```python
class InsufficientFundsError(Exception):
    pass

def withdraw(balance, amount):
    if amount > balance:
        raise InsufficientFundsError("Withdrawal amount exceeds available balance")
    return balance - amount

try:
    bal = 500
    amt = int(input("Enter withdrawal amount: "))
    new_balance = withdraw(bal, amt)
    print("Withdrawal successful. New balance:", new_balance)
except InsufficientFundsError as e:
    print(e)
```



**10.Implement a program that raises a custom exception if a user enters an incorrect password multiple times.
.**

```python
class TooManyAttemptsError(Exception):
    pass

def login(correct_password, max_attempts=3):
    attempts = 0
    while attempts < max_attempts:
        pwd = input("Enter password: ")
        if pwd == correct_password:
            print("Login successful")
            return
        attempts += 1
        print("Incorrect password")
    raise TooManyAttemptsError("Too many incorrect attempts")

try:
    login("python123")
except TooManyAttemptsError as e:
    print(e)
```
