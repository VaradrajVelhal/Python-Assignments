a given array whose sum equals a specific target number.**



```python
class PairSum:
    def find_pair(self, nums, target):
        for i in range(len(nums)):
            for j in range(i + 1, len(nums)):
                if nums[i] + nums[j] == target:
                    return i, j
        return None

arr = [2, 7, 11, 15]
target = 9

obj = PairSum()
result = obj.find_pair(arr, target)

if result:
    print("Pair found at indices:", result)
else:
    print("No pair found")
```

charge of any vehicle is seating capacity * 100. If Vehicle is a Bus instance, we
need to add an extra 10% on the full fare as a maintenance charge. So total fare
for bus instance will become the final amount = total fare + 10% of the total fare.**


```python
class Vehicle:
    def __init__(self, seating_capacity):
        self.seating_capacity = seating_capacity

    def fare(self):
        return self.seating_capacity * 100


class Bus(Vehicle):
    def fare(self):
        total_fare = super().fare()
        maintenance_charge = total_fare * 0.10
        final_fare = total_fare + maintenance_charge
        return final_fare


bus = Bus(50)

print("Total Bus Fare:", bus.fare())
```

```python
class RomanConverter:
    def int_to_roman(self, num):
        values = [1000, 900, 500, 400, 100, 90, 50, 40, 10, 9, 5, 4, 1]
        symbols = ["M", "CM", "D", "CD", "C", "XC", "L", "XL", "X", "IX", "V", "IV", "I"]

        roman = ""
        i = 0

        while num > 0:
            for _ in range(num // values[i]):
                roman += symbols[i]
                num -= values[i]
            i += 1

        return roman


obj = RomanConverter()
print(obj.int_to_roman(58))
```

']. These brackets must be close in the correct order, for example "()" and "()[]{}"
are valid but "[)", "({[)]" and "{{{" are invalid.**


```python
class ParenthesesValidator:
    def is_valid(self, s):
        stack = []
        brackets = {')': '(', '}': '{', ']': '['}

        for ch in s:
            if ch in brackets.values():
                stack.append(ch)
            elif ch in brackets:
                if not stack or stack.pop() != brackets[ch]:
                    return False

        return len(stack) == 0


obj = ParenthesesValidator()

print(obj.is_valid("()"))       
print(obj.is_valid("()[]{}"))   
print(obj.is_valid("({[)]"))   
print(obj.is_valid("{{{"))     
```

Animal with a method make_sound(). Create derived classes Dog and Cat that
override the method.**

```python
class Animal:
    def make_sound(self):
        print("Animal makes a sound")


class Dog(Animal):
    def make_sound(self):
        print("Dog barks")


class Cat(Animal):
    def make_sound(self):
        print("Cat meows")


dog = Dog()
cat = Cat()

dog.make_sound()
cat.make_sound()
```
