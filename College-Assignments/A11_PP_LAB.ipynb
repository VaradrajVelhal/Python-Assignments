**1.Write a Python program that reads a CSV file containing student names and their corresponding scores. Calculate the average score and print the names of students with scores above the average.**


```python
%%writefile students.csv
Name,Score
Dilip,78
Harsh,90
Harry,65
Tushar,88
Sudhir,75
Harshal,50
Gopi,60
Gopa,70
```

```python
import csv

filename = "students.csv"

try:
    scores = []
    students = []

    with open(filename, "r") as file:
        reader = csv.reader(file)

        header = next(reader)       # read header
        print("File Data:\n")
        print(header)               # print header row
        for row in reader:
            print(row)              # print each row (raw data)

            name = row[0]
            score = float(row[1])

            students.append((name, score))
            scores.append(score)

    # Calculate average
    avg = sum(scores) / len(scores)
    print("\nAverage score:", avg)

    # Students above avg
    print("\nStudents with score above average:")
    for name, score in students:
        if score > avg:
            print(name, "-", score)

except FileNotFoundError:
    print("Error: File", filename, "not found.")
```

```python
from datetime import datetime

filename = "notes.txt"

try:
   
    with open(filename, "r") as file:
        lines = file.readlines()  # list of lines

    print("Original file content:\n")
    for line in lines:
        print(line, end="")  
        # end="" prevents print() from adding an extra newline

    # adding timestamp to each line
    updated_lines = []
    for line in lines:
        timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        updated_lines.append(f"{timestamp} {line}") 

    
    with open(filename, "w") as file:
        file.writelines(updated_lines)

 
    print("\nUpdated file content :\n")
    with open(filename, "r") as file:
        print(file.read())

except FileNotFoundError:
    print("Error:", filename, "not found.")
```

```python
import json

filename = "books.json"

try:
    with open(filename, "r") as file:
        books = json.load(file)  #converts JSON into a Python list of dictionaries

    title = input("Enter book title to search: ").lower()

    found = False
    for book in books:
        if book.get("title", "").lower() == title:
            #If "title" does NOT exist returns "" , not an error
            print("\nBook found:")
            for key, value in book.items():
                print(f"{key}: {value}")
            found = True
            break

    if not found:
        print(f"{title} Book not found.")

except FileNotFoundError:
    print("Error: books.json file not found.")
except json.JSONDecodeError:
    print("Error: Invalid JSON format in file.")
```

**5.Write a Python program that creates a new directory and then creates multiple text files inside that directory, each with a different name and content..**


```python
import os

dir_name = "my_files"

# Create directory if it doesn't exist
os.makedirs(dir_name, exist_ok=True)

files_data = {
    "file.txt": "This is the content of file.",
    "file1.txt": "This is the content of file 1.",
    "file2.txt": "This is the content of file 2.",
}

# Create files and display their content
for fname, content in files_data.items():
    filepath = os.path.join(dir_name, fname)
    
    # Create and write content
    with open(filepath, "w") as file:
        file.write(content)

    # Display file name and content
    print(f"\nCreated file: {fname}")
    print("Content:")
    with open(filepath, "r") as file:
        print(file.read())

print("\nAll files created in directory:", dir_name)
```


**6.Create a program that reads a log file and extracts information about errors. Print the count of each type of error along with the corresponding error messages.**




```python
%%writefile log.txt
INFO: System started
ERROR: File not found
ERROR: Network failure
WARNING: Low memory
ERROR: File not found
ERROR Something went wrong

```

```python
filename = "log.txt"
error_counts = {} ## Dictionary to store error messages and how many times they appear


try:
    with open(filename, "r") as file:
        for line in file:
            line = line.strip() ## Remove extra spaces and newline characters
            if line.startswith("ERROR"):
               #  # Split the line into two parts at the first colon only
                # Example: "ERROR: File not found"  then ["ERROR", " File not found"]
                parts = line.split(":", 1)
                if len(parts) > 1:
                    message = parts[1].strip()
                else:
                    message = "Unknown error"
                #urrences of each error message
                error_counts[message] = error_counts.get(message, 0) + 1
                  # .get(message, 0) returns 0 if message does not exist yet


    if error_counts:
        print("Error summary:\n")
        for msg, count in error_counts.items():
            print(f"{msg} -> {count} time")
    else:
        print("No errors found in log file.")

except FileNotFoundError:
    print("Error: log file not found.")

```


**7.Create a program that reads a text file containing a list of email addresses. Identify and print all unique domains present in the email addresses.**


 

```python
%%writefile emails.txt
dilip@gmail.com
magar@yahoo.com
harsh@outlook.com
harry@gmail.com
sudhir@college.edu
student@pccoepune.org

```

```python
filename = "emails.txt"
domains = set()     # Using a set because sets store only unique values

try:
    # Open the file for reading only
    with open(filename, "r") as file:
        for line in file:

            email = line.strip()   # Remove spaces and newline at end

            # Only process valid emails that contain '@'
            if "@" in email:
                # Split email into username and domain
                # Example: "dilip@gmail.com" then ["dilip", "gmail.com"]
                _, domain = email.split("@", 1)

                domains.add(domain)   # Add domain to set duplicates ignored

    print("Unique domains found:\n")
    for d in domains:
        print(d)

except FileNotFoundError:
    print("Error: emails.txt not found.")
```

```python
input_file = "File_Handling.pdf"    
output_file = "File_Handling2.pdf"  
key = 0x55                            # XOR key 

try:
 
    with open(input_file, "rb") as infile, open(output_file, "wb") as outfile:

        data = infile.read()  
        
        # Encrypt using XOR:
        # XOR flips bits using the key. 
        # b ^ key means: take each byte from data and XOR it with key.
        encrypted = bytes([b ^ key for b in data])
        # We use bytes([...]) because XOR returns a list of integers,
        # and bytes() converts the list back into a byte sequence.

        outfile.write(encrypted)  
        
    print("Encryption done. Encrypted file:", output_file)


except FileNotFoundError:
    print("Error:", input_file, "not found.")


except IOError:
    print("I/O error during encryption.")
```

```python
import json
import os

filename = "employees.json"


if os.path.exists(filename):
    try:
        with open(filename, "r") as file:
            employees = json.load(file)  
            # json.load converts JSON data into a Python list
    except json.JSONDecodeError:
       
        employees = []
else:
    employees = []


print("Enter new employee details:")
emp_id = int(input("ID: "))       
name = input("Name: ")           
dept = input("Department: ")       

# Create a dictionary for the new employee
new_employee = {
    "id": emp_id,
    "name": name,
    "department": dept
}

# Add this dictionary to the list of employees
print("new employee details",new_employee)
employees.append(new_employee)

with open(filename, "w") as file:
    json.dump(employees, file, indent=4)
    # json.dump  writes list to file in JSON format

print("New employee added and data saved to", filename)

print("\nUpdated employees.json content:\n")

with open(filename, "r") as file:
    updated_data = json.load(file)
    print(json.dumps(updated_data, indent=4))
```
